```python
import pandas as pd
df = pd.read_csv("./input_data/iris.csv")
df.head()
```

```python
# 欠損値の確認
df.isnull().sum()
```

```python
df["data_version"] = 1.0
df.head()
```

```python
from sqlalchemy import create_engine
import os

# DBと接続するために必要な情報を辞書型でまとめる．
pgconfig = {
    "host" : os.environ["DB_HOST"],
    "port" : os.environ["DB_PORT"],
    "database" : os.environ["DB_NAME"], 
    "user": os.environ["DB_USER"], 
    "password" : os.environ["DB_PASSWORD"]
}

engine = create_engine("postgresql+psycopg2://{user}:{password}@{host}:{port}/{database}".format(**pgconfig))
    
```

```python
columns = ["data_version", 
           "sepal_length", 
           "sepal_width", 
           "petal_length", 
           "petal_width"]
'''
別データがすでにDBテーブル上にある時にどのように処理するかを指定する。
今回は、別データが存在することを想定しているので、データを加えるappendを選択。
既存データを削除して追加するreplaceなどのオプションもある

'''
df[columns].to_sql("iris_input", engine, if_exists="append", index = False)
```

## データの参照
- データの読み込みについてはipython-sqlを使う

```python
%load_ext sql
dsl = "postgres://{user}:{password}@{host}:{port}/{database}".format(**pgconfig)
%sql $dsl
```

```python
%config SqlMagic.autopandas = True # Trueにすることで、処理結果を直接pandasのdataframeとして読み込める
df_input = %sql select * from iris_input
df_input.head()
```

- 以上で、DBサーバーにデータを保存することができた。

## 分析の実行
- DBから取得したデータに対して簡単なクラスタリングを行う．

```python
df_1 = df_input[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
df_1.head()
```

```python
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters = 3, random_state = 0)   # 3つのクラスに分けるクラスタリングを行う
clusters = k_means.fit_predict(df_1)
clusters
```

## データの可視化をするよ

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2, random_state = 0)  # 主成分の数を2に設定.
pca_result = pca.fit_transform(df_1)      # array型
df_2 = pd.DataFrame(pca_result, columns=["pca_1", "pca_2"])  # データフレーム型に変更する
df_2.head()
```

```python
df_2 = df_2.assign(cluster_id = clusters)
df_2.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("seaborn")
%matplotlib inline

# 線形回帰はいらないのでfit_reg = False
graph = sns.lmplot(x = "pca_1", y = "pca_2", data = df_2, hue = "cluster_id", markers = ["o", "s", "+"], fit_reg = False)
graph.fig.set_figheight(10)
graph.fig.set_figwidth(10)
```

## 分析結果をDBに保存

```python
 df_input.head()
```

- 元々のデータには管理しやすいように、idとdata_versionが設定されていた。
- そこで、分析結果にもこのカラムを入れる

```python
df_save = df_2.assign(id = df_input[["id"]], data_version = df_input[["data_version"]])
df_save.head()
```

- DataFrameのto_sqlにてDBに保存する

```python
df_save.to_sql('iris_output', engine, if_exists = "append",  index = False)
```
